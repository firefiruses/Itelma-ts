```python
import pandas as pd
import os
```

```python
def correct_folder(nbr, folder, class_name, delimeter):
    rst = []
    for x in str(nbr).split(delimeter):
        if str(x) in os.listdir(os.path.join(folder, class_name)):
            rst.append(x)
        else:
            print(x)
            print(sorted(os.listdir(os.path.join(folder, class_name)), key=int))
    if len(rst):
        return delimeter.join(rst)
    return None
```

```python
sick_info = pd.read_excel("itelma_dataset/hypoxia.xlsx")["folder_id"].iloc[1:].reset_index(drop=True)
sick_info = pd.DataFrame(data=sick_info, columns=["folder_id"])
sick_info["has_hypoxia"] = True
sick_info["folder_id"] = sick_info["folder_id"].apply(lambda x: correct_folder(x, "itelma_dataset", "hypoxia", ","))
sick_info = sick_info[sick_info["folder_id"].notna()]
sick_info
```

```python
fine_info = pd.read_excel("itelma_dataset/regular.xlsx", header=1)["Folder_id"].dropna().apply(lambda x: str(x).replace(";", ",")).reset_index(drop=True)
fine_info = pd.DataFrame(data=fine_info.to_numpy(), columns=["folder_id"])
fine_info["has_hypoxia"] = False
fine_info["folder_id"] = fine_info["folder_id"].apply(lambda x: correct_folder(x, "itelma_dataset", "regular", ","))
fine_info = fine_info[fine_info["folder_id"].notna()]
fine_info
```

```python
EPOCHS = 50
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
#loss_fn = nn.CrossEntropyLoss(weight=torch.tensor([.75, 1.25], device=device))
loss_fn = nn.L1Loss()
optimizer = torch.optim.NAdam(model.parameters())
optimizer.zero_grad()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
import warnings; warnings.filterwarnings("ignore")


for i in range(EPOCHS):
    model.train()
    loss_sum = 0
    sample_num = 0
    for _x_batch, y_batch in dl:
        #_x_batch = {k: v.to(device) for k, v in _x_batch.items()}
        #print(_x_batch["folders"], y_batch)
        y_batch = y_batch.to(device)
        x_batch_delta = torch.cat([torch.tensor([[0.0] for _ in range(len(y_batch))]), _x_batch["bpm_ts"][:, 1:] - _x_batch["bpm_ts"][:, :-1]], dim=1)
        t = int(torch.randint(100, 1000, (1,))[0])
        a = int(torch.randint(0, _x_batch["bpm"].shape[1] - t - 2, (1,))[0])
        x_batch = torch.stack([_x_batch["bpm"], _x_batch["bpm_ts"], x_batch_delta], dim=2)[..., a:a+t+2, :].contiguous().to(device)
        y_pred = model(x_batch[..., :-1, :])
        #print(y_batch.shape, y_pred.shape)
        loss = loss_fn(y_pred, x_batch[..., -1, :-2])
        loss.backward()
        loss_sum += loss.item()
        sample_num += len(y_batch)
        optimizer.step()
        optimizer.zero_grad()
    print(f"Avg loss on epoch {i + 1}/{EPOCHS} is {loss_sum / sample_num}")
    model.eval()
    loss_sum = 0
    sample_num = 0
    _y_pred = []
    _y_batch = []
    with torch.no_grad():
        for _x_batch, y_batch in dl_val:
            #_x_batch = {k: v.to(device) for k, v in _x_batch.items()}
            y_batch = y_batch.to(device)
            x_batch_delta = torch.cat([torch.tensor([[0.0] for _ in range(len(y_batch))]), _x_batch["bpm_ts"][:, 1:] - _x_batch["bpm_ts"][:, :-1]], dim=1)
            x_batch = torch.stack([_x_batch["bpm"], _x_batch["bpm_ts"], x_batch_delta], dim=2)[..., -501:, :].contiguous().to(device)
            y_pred = model(x_batch[..., :-1, :])
            #print(y_batch, y_pred)
            loss = loss_fn(y_pred, x_batch[..., -1, :-2])
            #loss.backward()
            loss_sum += loss.item()
            sample_num += len(y_batch)
            #optimizer.step()
            #optimizer.zero_grad()
            #_y_pred.extend(torch.argmax(y_pred, dim=1).tolist())
            #_y_batch.extend(torch.argmax(y_batch, dim=1).tolist())
    print(f"Avg loss on epoch {i + 1}/{EPOCHS} is {loss_sum / sample_num}")
    #print(f"Accuracy: {accuracy_score(_y_batch, _y_pred)}\nPrecision: {precision_score(_y_batch, _y_pred)}\nRecall: {recall_score(_y_batch, _y_pred)}\n")
```
